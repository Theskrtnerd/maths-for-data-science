First we import some packages we need

```python
import numpy as np
import matplotlib.pyplot as plt
```

Now we will define some functions

```python
def g(x):
    return (x+1)**2
def h(x):
    return x**(1/2)-1
```

(ùëé) We define $p = g \cdot h$

```python
def p(x):
    return g(x)*h(x)
```

Now we plot it. We need to notice that $h(x)$ will only be defined for $x \le 0$. After some experimenting I find that a range which shows the interesting features of the plot.

```python
x = np.arange(0,10,0.1)
plt.plot(x,p(x))
```

Now we compose them

(ùëè) We define $r(x) = g(h(x))$

```python
def r(x):
    return g(h(x))
```

Something important to note: $h(x)$ has the domain of $[0, +\infty)$. So $r(x)$ will also has the domain of $[0, +\infty)$.

Now we will plot $r(x)$.

```python
t = np.arange(0,10,0.1)
plt.plot(t,r(t))
```

Side note: Notice that $r(x) = x$. This is indeed true since $r(x) = g(h(x)) = g(\sqrt{x} ‚àí 1) =
(\sqrt{x} ‚àí 1 + 1)^2 = x$

(ùëê) We define $s(x) = h(g(x))$

```python
def s(x):
    return h(g(x))
```

Something important to note: $g(x) \ge 0$, for all $x$. So $s(x)$ is defined for all $x$. In order to get a proper picture of it we will need to include some negative values in the domain.

```python
k = np.arange(-2,2,0.1)
plt.plot(k,s(k))
```
