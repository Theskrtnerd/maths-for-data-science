We can generalize the series as:
$$
S = \sum_{i=1}^{\infty} (-1)^{i-1} \cdot \frac{1}{3i-2}
$$

Let’s denote $S_n$ as
$$
S_n = \sum_{i=1}^{n} (-1)^{i-1} \cdot \frac{1}{3i-2}
$$


So with this notation, $S_1=1$, $S_2 = 1 - \frac{1}{4} = \frac{3}{4}$, and so on.
Notice that as $n$ gets larger, $S_n$ will get closer to $S$.

```python
def S(m,n): #This function will print the S_i for i between m and n
    s = 0
    for i in range(1,n+1):
        a = (-1)**(i-1) * 1/(i*3-2)
        s = s+ a
        if i >= m:
            print(f'S_{i} is equal to {s}')
S(5003,5005)
S(10000,10000)
S(100000,100000)
S(1000000,1000000)
S(10000000,10000000)
```

So notice that the 4 decimal places of $S_n$ don’t change after $n = 5003$. So it’s clear that the answer of S, accurate to 4 decimal places is 0.8356.

We have the Taylor series of $sin(x)$ is:
$$
sin(x) = x - \frac{x^3}{3!} + \frac{x^5}{5!} - ...
$$

Hence, we have:
$$
P_3(x) = x - \frac{x^3}{3!}
$$
$$
P_5(x) = x - \frac{x^5}{5!}
$$

We will define all the functions $sin(x)$, $P_3(x)$, $P_5(x)$ below.

```python
import numpy as np
def sin(x):
    return np.sin(x)
def P_3(x):
    return x - x**3/6
def P_5(x):
    return x - x**3/6 + x**5/120
```

And now we will plot them, with a domain of approximately $[-\pi, \pi]$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 1000)

plt.figure(figsize=(10, 6))

plt.plot(x, sin(x), label='sin(x)')
plt.plot(x, P_3(x), label='P_3(x)')
plt.plot(x, P_5(x), label='P_5(x)')

plt.legend()
plt.show()
```
